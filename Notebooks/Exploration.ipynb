```python

tech_tols = ["engine", "torque", "mileage", "max_power", "seats"]
df = df.dropna(subset = tech_tols)
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns=["name", "seller_type", "owner"], errors = "ignore")
df
```

```python
df["mileage_suffix"] = (
    df["mileage"]
    .astype(str)
    .str.replace(r"[\d\.]+", "", regex=True)
    .str.strip()
    .str.lower()
)

df["mileage_suffix"].value_counts()
```

```python
df["engine_suffix"] = (df["engine"].astype(str).str.replace(r"[\d\.]+", "", regex = True).str.strip().str.lower())
df["engine_suffix"].value_counts()
```

```python
df["max_power_suffix"] = (df["max_power"].astype(str).str.replace(r"[\d\.]+", "", regex = True).str.strip().str.lower())
df["max_power_suffix"].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["selling_price"], bins = 150)
plt.ticklabel_format(style = "plain", axis = "x")
plt.xlim(0,3_000_000)
plt.xlabel("Giá xe")
plt.ylabel("Số xe")
plt.title("Phân phối xe theo giá")
plt.show()
```

```python
df["selling_price"].describe()
```

```python
priceByFuel = df.groupby("fuel")["selling_price"].agg(mean_price ="mean", median_price = "median", count = "count").reset_index()
priceByFuel
```

```python
plt.figure(figsize=(8,5))
plt.bar(priceByFuel["fuel"], priceByFuel["mean_price"])
plt.xlabel("Loại nhiên liệu")
plt.ylabel("Giá trung bình")
plt.title("Giá xe trung bình theo nhiên liệu")
plt.ticklabel_format(style = "plain", axis = "y")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(priceByFuel["fuel"], priceByFuel["median_price"])
plt.xlabel("Loại nhiên liệu")
plt.ylabel("Giá trung vị")
plt.title("Giá xe trung vị theo loại nhiên liệu")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
currentYear = 2025
df["car_age"] = currentYear - df["year"]
df[["year", "car_age"]].head()
```

```python
plt.figure(figsize = (8,5))
plt.scatter(df["car_age"], df["selling_price"], alpha = 0.4)
plt.xlabel("Tuổi của xe")
plt.ylabel("Giá của xe")
plt.title("Mối quan hệ giữa giá và tuổi của xe")
plt.ticklabel_format(style = "plain", axis = "y")
plt.show()
```

```python
corr = df["car_age"].corr(df["selling_price"])
corr
```

```python
priceByAge = df.groupby("car_age")["selling_price"].mean().reset_index().sort_values("car_age")
priceByAge
```

```python
df["priceByAge"] = df.groupby("car_age")["selling_price"].transform("mean")
df.head()
```

```python
plt.figure(figsize = (9,5))
plt.plot(priceByAge["car_age"], priceByAge["selling_price"], marker = "o")
plt.xlabel("Tuổi của xe")
plt.ylabel("Giá bán trung bình")
plt.title("Xu hướng khấu hao của xe theo năm sử dụng")
plt.ticklabel_format(style = "plain", axis = "y")
plt.show()
```
